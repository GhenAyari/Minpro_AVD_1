###outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

###construct data dan penambahan column

```python
#disini saya menambahkan column total cases = jumlah total kasus (underweight + healthyWeight + overweight)
data = {
    "SchoolYear": [2020, 2020, 2021, 2021],
    "NameHospital": ["Hospital A", "Hospital B", "Hospital A", "Hospital B"],
    "Sex": ["Male", "Female", "Male", "Female"],
    "EpiUnderweight": [50, 30, 40, 20],
    "EpiHealthyWeight": [300, 250, 280, 260],
    "EpiOverweight": [80, 70, 90, 60],
}

df = pd.DataFrame(data)

df["TotalCases"] = df["EpiUnderweight"] + df["EpiHealthyWeight"] + df["EpiOverweight"]
```

###Visualisasi

```python
cases_per_year = df.groupby('SchoolYear')['TotalCases'].sum()
plt.figure(figsize=(10, 6))
cases_per_year.plot(kind='bar', color='blue')
plt.title('Jumlah Kasus per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Dari hasil visualisasi jumlah kasus per tahun terlihat bahwa pada tahun 2020 lebih besar dari 2021

```python
plt.figure(figsize=(6, 4))
plt.bar(df["Sex"], df["TotalCases"], color=["blue", "pink"])
plt.title("Jumlah Kasus Berdasarkan Jenis Kelamin")
plt.xlabel("Jenis Kelamin")
plt.ylabel("Jumlah Kasus")
plt.show()
```

Dari visualisasi, terlihat bahwa jumlah total kasus pada laki-laki lebih tinggi dibandingkan perempuan.
Ini menunjukkan bahwa anak laki-laki lebih banyak mengalami masalah berat badan (underweight, healthy weight, overweight) dibandingkan anak perempuan.

Kemungkinan Penyebab Perbedaan, Laki-laki cenderung memiliki kebutuhan energi lebih tinggi dibandingkan perempuan, sehingga mereka lebih rentan mengalami overweight jika pola makan tidak terkontrol atau underweight jika asupan gizi tidak mencukupi.
Gaya Hidup dan Aktivitas Fisik: Anak laki-laki mungkin lebih aktif secara fisik, tetapi juga lebih rentan terhadap pola makan yang kurang teratur, yang bisa menyebabkan ketidakseimbangan berat badan.

```python
tahun_tertentu = 2020
df_tahun = df[df["SchoolYear"] == tahun_tertentu]
total_underweight = df_tahun["EpiUnderweight"].sum()
total_healthy = df_tahun["EpiHealthyWeight"].sum()
total_overweight = df_tahun["EpiOverweight"].sum()
labels = ["Underweight", "Healthy Weight", "Overweight"]
sizes = [total_underweight, total_healthy, total_overweight]
colors = ["lightcoral", "lightblue", "lightgreen"]
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90)
plt.title(f"Persentase Kategori Berat Badan Tahun {tahun_tertentu}")
plt.show()
```

Sebagian besar anak memiliki berat badan sehat

82,6% anak berada dalam kategori Healthy Weight, yang berarti mayoritas anak memiliki berat badan yang sesuai dengan standar kesehatan.
Ini menunjukkan bahwa pola makan dan gaya hidup anak-anak secara umum masih cukup baik dalam menjaga keseimbangan berat badan.
Persentase Anak Overweight Cukup Tinggi

16,5% anak mengalami overweight, yang berarti ada kecenderungan sebagian anak mengalami kelebihan berat badan yang bisa berisiko menjadi obesitas jika tidak dikontrol.
Kenaikan angka overweight ini bisa menjadi perhatian khusus, terutama jika terus meningkat dari tahun ke tahun.
Underweight Sangat Rendah

Hanya 0,9% anak yang berada dalam kategori underweight, menunjukkan bahwa kasus kekurangan berat badan sangat sedikit dibandingkan overweight.
Ini bisa berarti bahwa akses ke makanan atau pola makan yang kurang tidak menjadi masalah besar dibandingkan risiko kelebihan berat badan.

```python
tahun_tertentu = 2021
df_tahun = df[df["SchoolYear"] == tahun_tertentu]
total_underweight = df_tahun["EpiUnderweight"].sum()
total_healthy = df_tahun["EpiHealthyWeight"].sum()
total_overweight = df_tahun["EpiOverweight"].sum()
labels = ["Underweight", "Healthy Weight", "Overweight"]
sizes = [total_underweight, total_healthy, total_overweight]
colors = ["lightcoral", "lightblue", "lightgreen"]
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90)
plt.title(f"Persentase Kategori Berat Badan Tahun {tahun_tertentu}")
plt.show()
```

Mayoritas anak masih dalam Kategori berat badan sehat

Penurunan Persentase Anak Overweight

Overweight menurun menjadi 14,1% dari 16,5% di tahun 2020, yang berarti lebih sedikit anak yang mengalami kelebihan berat badan.
Ini bisa menjadi indikasi bahwa ada perbaikan dalam pola makan atau peningkatan aktivitas fisik di kalangan anak-anak.
Sedikit Peningkatan Kasus Underweight

Underweight meningkat dari 0,9% (2020) menjadi 1,3% (2021), meskipun masih dalam jumlah yang sangat kecil.
Ini bisa menunjukkan bahwa meskipun overweight menurun, ada sedikit peningkatan jumlah anak yang mengalami kekurangan berat badan.
